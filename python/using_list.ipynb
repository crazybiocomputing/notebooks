### 1.0. Initialization

Run the following code and then answer the questions (Q1, Q2, etc).

```
#@title Run this script first for creating the list _arr_ then answer the questions
import random
import math

def generate():
  length = random.randint(30,110)
  arr = [0] * length
  for i,v in enumerate(arr):
    arr[i] = random.randint(0,300)
  return arr

arr = generate()

```

### 1.4. Question #4  ★☆☆

Q4: What is the sum of all the elements in the list **arr**?

```
# Add some code here
```

```
#@title Q4 - Run it and check your answer
import functools
import ipywidgets as widgets
from IPython.display import display

field = widgets.Text(description="Q4: ",value = '0')
button = widgets.Button(description="Check")
label = widgets.HTML(value = f"<b><font color='gray'>??</b>")
hbox = widgets.HBox([ field,button,label])
output = widgets.Output()

display(hbox, output)

def on_button_clicked(b):
  with output:
    _sum = functools.reduce(lambda accu,b: accu+b,arr,0)
    if field.value == str(_sum):
      label.value = f"<b><font color='green'>Correct!</b>"
    else:
      label.value = f"<b><font color='red'>Incorrect!</b>"     
button.on_click(on_button_clicked)
```
